```python
from sklearn.model_selection import train_test_split
```

```python
health.info()
```

```python
X = health[[
    'Employer Health Insurance Coverage (2015)',  # Employer-based coverage
    'Marketplace Health Insurance Coverage (2016)',  # ACA marketplace enrollment
    'Medicaid Enrollment (2016)',  # Public insurance coverage
    'Medicare Enrollment (2016)',  # Elderly coverage
    'Average Monthly Tax Credit (2016)'  # Subsidy affordability
]]
y = health['Uninsured Rate (2015)'].str.rstrip('%').astype(float)  # Convert to numeric
```

```python
X.head()
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train, y_train)
```

```python
lm.coef_
```

```python
lm.score(X, y)
```

```python
cdf = pd.DataFrame(lm.coef_,X.columns,columns=['Coef'])
print(cdf)
```

```python
import statsmodels.api as sm
```

```python
X = sm.add_constant(X_train)
model = sm.OLS(y_train, X)
model_fit = model.fit()
print(model_fit.summary())
```

```python
predictions = lm.predict(X_test)
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=y_test, y=predictions)
plt.ylabel('Predicted Uninsured Rate')
plt.title('Uninsured Rate: Actual vs Model Predictions')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import math
```

```python
print('Mean Absolute Error:',mean_absolute_error(y_test, predictions))
print('Mean Squared Error:',mean_squared_error(y_test, predictions))
print('Root Mean Squared Error:',math.sqrt(mean_squared_error(y_test, predictions)))
```

```python
residuals = y_test-predictions
sns.distplot(residuals, bins=30)
```

```python
import pylab 
import scipy.stats as stats

stats.probplot(residuals, dist="norm", plot=pylab)
pylab.show()
```

#This dataset shows how health insurance coverage changed across U.S. states from 2010 to 2016. The main goal was to predict the uninsured rate in 2015 using factors like Medicaid enrollment, employer-based insurance, and government subsidies. The probability plot (Q-Q plot) checks if the model’s errors follow a normal pattern—if the points mostly line up along the diagonal, the model works well; if they curve away, the predictions might be off for very high or low uninsured rates.

From the data, we can see that states with expanded Medicaid programs and more people enrolled in marketplace insurance (Obamacare) tended to have lower uninsured rates. Employer coverage also played a big role—where it was high, fewer people were uninsured. A few states had missing data, which could affect results, and some percentages needed cleaning (like removing the "%" sign).

To improve the model, we might need to adjust for outliers, add more details (like state poverty rates), or try different prediction methods. Overall, the data helps us understand what policies (like Medicaid expansion) most reduce uninsured rates—useful for future healthcare planning.
